```python
import warnings 
warnings.filterwarnings("ignore")
#to filter the massage show in collab

import requests #call requests library
from bs4 import BeautifulSoup #call BeautifulSoup library 

from urllib.request import Request, urlopen
import urllib 
import urllib.error  # to use it withh error when access website browser
import time  #to display data in time date expresion 
#from lPython.display import YouTubeVideo  #run youtube file 

user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
#find user agent and assign it in a variable 
headers = {'user-Agent':user_agent} #create dictinary header and store user agent   

params = {
    "view_op":"view_org" ,    #author result
    "org":'8607118205373147890' ,   #orgnization ID 
    "hl":"en" ,     #language 
}     #from url in browser page up
```

```python
url = 'https://scholar.google.com/citations?'

try:
    reqest = requests.get(url,headers=headers, params=params)
except urllib.error.HTTPError as e:
    # Return code error (e.g. 404, 501, ...)
    # ...
    print('HTTPError: {}'.format(e.code))
except urllib.error.URLError as e:
    # Not an HTTP-specific error (e.g. connection refused)
    # ...
    print('URLError: {}'.format(e.reason))
else:
    # 200
    # ...
    print('The requested page is rendered successfully: the response status is good')
```

```python
def google_citation_by_university(page_url)->BeautifulSoup:
    """Get html content of first page of a given university.
    Parameters
    ----------
    org : str
        Organization ID given by Google.
    Returns
    -------
    page_cont : BeautifulSoup object
        Html content of the page.
    """
    url = page_url
    reqest = Request(url,headers=headers)
    page = urlopen(reqest)
    page_cont = BeautifulSoup(page, "lxml")
    return page_cont
```

```python
new_http='https://scholar.google.com/citations?view_op=view_org&hl=en&org=8607118205373147890' 
from selenium import webdriver
driver = webdriver.Chrome()
employee1 = []
while(new_http!=''):
    driver.get(new_http)
    if(driver.find_element(By.CLASS_NAME ,'gs_btnPR').is_enabled() == False):
        new_http=""
        pass
    else:
        #find element by partial link text
        myLink = driver.find_element(By.CLASS_NAME ,'gs_btnPR')
        myLink.click()
        new_http = driver.current_url
        univ_page= google_citation_by_university(new_http)
        div_name = univ_page.find_all(class_= "gs_ai_name")
        div_cited= univ_page.find_all(class_="gs_ai_cby")
        div_aff = univ_page.find_all(class_="gs_ai_aff")
        div_int = univ_page.find_all(class_="gs_ai_int")
        div_url = driver.find_element(By.XPATH ,'/html/body/div/div[7]/div[2]/div/div[2]/div/div/h3/a').get_attribute('href')
        for i , j , n ,t in zip(div_name , div_cited ,div_aff , div_int )  :
            employee1.append({"name":i.text , "citization":j.text , "aff" : n.text , "interesting " : t.text , "link" :div_url })
        
```
